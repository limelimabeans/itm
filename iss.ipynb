```python
%matplotlib inline
import requests as rq
import json
import time
import numpy as np
import matplotlib.pyplot as plt

longitudes = []
latitudes = []

for i in range(100):
    response = rq.get('http://api.open-notify.org/iss-now.json').text
    response_info = json.loads(response)
    longitudes.append(float(response_info['iss_position']['longitude']))
    latitudes.append(float(response_info['iss_position']['latitude']))
    time.sleep(10)

plt.figure(figsize=(10, 6))
plt.plot(longitudes, latitudes, 'b.-')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('International Space Station Path')
plt.grid(True)
plt.show()
```
